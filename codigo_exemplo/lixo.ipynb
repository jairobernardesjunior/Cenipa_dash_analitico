```python
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import warnings
warnings.simplefilter('ignore')
```

```python
dfx = pd.read_csv('../dataset_analise/df_acidentes_analise_ocorr.csv')
```

```python
dfx['ocorrencia_latitude'] = dfx['ocorrencia_latitude'][( \
                        (dfx['ocorrencia_uf']=='AM') | \
                        (dfx['ocorrencia_uf']=='RR') | \
                        (dfx['ocorrencia_uf']=='PB')) & \
                        (dfx['ocorrencia_latitude'] > 0) & \
                        (dfx['ocorrencia_latitude'] != 'NaN') \
                        ] + 23.227222

dfx['ocorrencia_longitude'] = dfx['ocorrencia_longitude'][( \
                        (dfx['ocorrencia_uf']=='AM') | \
                        (dfx['ocorrencia_uf']=='RR') | \
                        (dfx['ocorrencia_uf']=='PB')) & \
                        (dfx['ocorrencia_longitude'] > 0) & \
                        (dfx['ocorrencia_longitude'] != 'NaN') \
                        ] + 3.172777
```
